## 1. Write a Python programming to display a bar chart of the popularity of programming Languages.
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array(["Java","Python","PHP","Javascript","C#","C++"])
y=np.array([22.2,17.6,8.8,8,7.7,6.7])
plt.bar(x,y,color="m")
plt.grid(which="major")
plt.grid(which="minor")
plt.title("Popularity of programming languages world wide")
plt.xlabel("Languages")
plt.ylabel("Popularity")

plt.show()
```

## 3. Write a Python program to display a bar chart of the popularity of programming Languages. Use uniform color.
